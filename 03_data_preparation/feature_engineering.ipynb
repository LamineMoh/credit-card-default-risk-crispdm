# Feature Engineering

## Objective
Create meaningful features based on financial behavior and payment history to improve model performance.


```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\zouta\Desktop\Projets\Default Of Credit Cards Clients\data\processed\credit_default_cleaned.csv")
df.shape
```

```python
df.columns
```

```python
pay_cols = ["PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]

df["avg_pay_delay"] = df[pay_cols].mean(axis=1)
df["max_pay_delay"] = df[pay_cols].max(axis=1)
```

```python
bill_cols = ["BILL_AMT1", "BILL_AMT2", "BILL_AMT3",
             "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]

df["avg_bill_amt"] = df[bill_cols].mean(axis=1)
df["credit_utilization"] = df["avg_bill_amt"] / df["LIMIT_BAL"]
```

```python
pay_amt_cols = ["PAY_AMT1", "PAY_AMT2", "PAY_AMT3",
                "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]

df["avg_payment_amount"] = df[pay_amt_cols].mean(axis=1)
```
